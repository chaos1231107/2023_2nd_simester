### Data Cleansing and Preprocessing

```python
c = list(dch.columns)
print('Before: ', c[0:3])
for i, ci in enumerate(c):
    c[i] = ci.strip()
print('After: ', c[0:3])
dch.columns = c
dch = dch[dch['Trip Distance (miles)'] > 1e-5]
#shift start to zero
dch['Time (sec)'] = dch['Time (sec)'] - dch['Time (sec)'].iloc[0]
#filter based on
dch = dch[dch['Bearing (deg)'] >= 0]
#every 10th row
dch = dch[::10]
#set index
dch.set_index('Time (sec)', inplace=True)
dch
```

### Average Fuel Economy

```python
a = dch['Trip Distance (miles)']
b = data['Trip Fuel (gal)']
dch['Avg fuel economy (MPG)'] = a / b

t1 = np.array(data.index) / 60 #Ford
t2 = np.array(dch.index) / 60 #Crysler

plt.figure(figsize=(10,8))
plt.subplot(4,1,1)
plt.plot(t1,data['Vehicle speed (MPH)'].values,'r-', label='Ford')
plt.plot(t2,dch['Vehicle speed (MPH)'].values, 'b--', label='Chrsler')
plt.grid(); plt.ylabel('Speed (mph)')
plt.legend()

plt.subplot(4,1,2)
plt.plot(t1, data['Trip Distance (miles)'].values, 'r-', label='Ford')
plt.plot(t2, dch['Trip Distance (miles)'].values, 'b--', label='Chysler')
plt.grid(); plt.ylabel('Distance (mi)')
plt.legend()


plt.show()
```
