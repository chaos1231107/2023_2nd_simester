```python
#practice 3
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import tensorflow as tf
```

```python
#numpy
def theta(t):
    return 0 + 0.5*np.pi*(t**2)

def position(L, theta):
    return L*np.cos(theta), L*np.sin(theta)
```

```python
t = np.linspace(0,2,100)
L = 1.5
p_val = position(L, theta(t))
np.array(p_val)
```

```python
plt.plot(t, p_val[0], 'r')
plt.plot(t, p_val[1], 'g')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(p_val[0], p_val[1],'ko')
plt.show()
```

```python
#sympy
L, t = sp.symbols('L, t')
theta = 0 + 0.5*np.pi*(t**2)
theta
```

```python
position_sp = (L*sp.cos(theta), L*sp.sin(theta))
position_sp
```

```python
t_val = np.linspace(0,2,100)

pos_val = []
for pos in position_sp:
    pos_val_dummy = []
    for t_unit in t_val:
        pos_val_dummy.append(pos.subs([(L, 1.5), (t, t_unit)]))
    pos_val.append(pos_val_dummy)
```

```python
np.array(pos_val)
```

```python
plt.figure(figsize=(5,5))
plt.plot(pos_val[0], pos_val[1], 'ko')
plt.show()
```

```python
L, t = sp.symbols('L, t')
omega = np.pi*t
theta = sp.integrate(omega, t)
theta
```

```python
theta_const = -theta.subs(t,0) + 0
theta_const
```

```python
theta += theta_const
theta
```

```python
position_sp = (L*sp.cos(theta), L*sp.sin(theta))
t_val = np.linspace(0,2,100)

pos_val = []
for t_unit in t_val:
    pos_dummy = []
    for pos in position_sp:
        pos_dummy.append(pos.subs([(L, 0.5), (t, t_unit)]))
    pos_val.append(pos_dummy)
pos_val = np.array(pos_val)
```

```python
plt.figure(figsize=(5,5))
plt.plot(pos_val[:,0], pos_val[:,1],'ko')
```
